**Plotting the Sigmoid Function**

```python
X_exp = np.linspace(-4,4,81)
print(X_exp[:5])
print(X_exp[-5:])
```

```python
Y_exp = np.exp(X_exp)
Y_exp[:5]
```

```python
plt.plot(X_exp, Y_exp)
plt.title('Plot of $e^X$')
```

```python
Y_exp = np.exp(-X_exp)
plt.plot(X_exp, Y_exp)
plt.title('Plot of $e^{-X}$')
```

```python
def sigmoid(X):
    Y = 1 / (1 + np.exp(-X))
    return Y
```

```python
X_sig = np.linspace(-7,7,141)
Y_sig = sigmoid(X_sig)
plt.plot(X_sig,Y_sig)
plt.yticks(np.linspace(0,1,11))
plt.grid()
plt.title('The sigmoid function')
```
